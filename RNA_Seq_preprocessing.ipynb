```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores = []
K = range(2, 6)  # Test 2 to 5 clusters
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, clusters)
    silhouette_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")

# Plot silhouette scores
import matplotlib.pyplot as plt
plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```

```python


# 1) Load metadata
# (adjust the paths/sep if yours are TSV or CSV)
luad = pd.read_csv("./luad_clinical_sample.txt", sep="\t", comment="#", skiprows=3)
lusc = pd.read_csv("./lusc_clinical_sample.txt", sep="\t", comment="#", skiprows=3)

# 2) Pick out just the Sample IDs and assign labels
#    Here we'll encode LUAD as 1, LUSC as 0
luad = luad[["SAMPLE_ID"]].copy()
luad["label"] = 1
lusc = lusc[["SAMPLE_ID"]].copy()
lusc["label"] = 0

print(luad.head())
print(lusc.head())
```

```python
# 3) Combine into one DataFrame
meta = pd.concat([luad, lusc], ignore_index=True)
meta["SAMPLE_ID"] = meta["SAMPLE_ID"].str.replace(r"-01$", "", regex=True)
meta.head()
```

```python
# 4) Load your expression matrix
#    rows = genes, columns = sample barcodes (e.g. TCGA-05-4244-01, TCGA-18-3406-01, …)
expr = df_RNASeq[df_RNASeq.columns[2:]]

# 5) Sometimes the metadata SAMPLE_ID is 12‑chars (TCGA‑05‑4244)  
#    while your expr columns are 15‑chars (TCGA‑05‑4244‑01).  
#    If so, truncate your expr column names to the first 12 chars:
expr_cols_12 = expr.columns.str.slice(0,12)
expr.columns = expr_cols_12

# 6) Keep only the samples for which we have labels
common = expr.columns.intersection(meta["SAMPLE_ID"])
expr_sub = expr[common]

# 7) Build X (features) and y (labels)
X = expr_sub.T            # shape = (n_samples, n_genes)
y = meta.set_index("SAMPLE_ID").loc[common, "label"]

print("X:", X.shape)
print("y distribution:\n", y.value_counts())
```

```python
from lime.lime_tabular import LimeTabularExplainer

gene_names = final_data['Hugo_Symbol'].tolist()  # length 994

# ── B) Create the LIME explainer ──────────────────────────────────────────────
explainer_lime = LimeTabularExplainer(
    training_data   = X_train_scaled,
    feature_names   = gene_names,
    class_names     = ['LUSC', 'LUAD'],
    mode            = 'classification'
)

# ── C) Explain one instance (local explanation) ───────────────────────────────
i = 0  # index of the test sample you want to inspect
exp = explainer_lime.explain_instance(
    data_row    = X_test_scaled[i],
    predict_fn  = model.predict,     # your Keras model returns probability vector
    num_features= 10                 # top 10 genes for this sample
)

# If you want a bar chart for this single sample:
feat, weight = zip(*exp.as_list(label=1))
plt.figure(figsize=(6,4))
plt.barh(feat[::-1], weight[::-1], color='coral')
plt.title(f"LIME Local Explanation (sample #{i})")
plt.xlabel("LIME weight")
plt.tight_layout()
plt.show()
```

```python
n_agg = 50  # number of random test samples to aggregate
agg_weights = np.zeros(len(gene_names))

# For each random sample, get LIME weights for class 1
idxs = np.random.choice(range(X_test_scaled.shape[0]), n_agg, replace=False)
for j in idxs:
    exp_j = explainer_lime.explain_instance(
        X_test_scaled[j],
        model.predict,
        num_features=len(gene_names)   # get all gene weights
    )
    # as_list returns only the top num_features; instead grab .as_map()
    # .as_map()[class_id] is list of (feature_index, weight)
    local_map = dict(exp_j.as_map()[1])
    for feat_idx, w in local_map.items():
        agg_weights[feat_idx] += abs(w)

# average
agg_weights /= n_agg

# pick top 10 genes
top_idx   = np.argsort(agg_weights)[-10:][::-1]
top_genes = [gene_names[i] for i in top_idx]
top_vals  = agg_weights[top_idx]

# ── E) Plot global top‑10 bar chart ────────────────────────────────────────────
plt.figure(figsize=(8,5))
plt.barh(top_genes[::-1], top_vals[::-1], color='teal')
plt.xlabel("Mean |LIME weight| over 50 samples")
plt.title("Top 10 Genes by LIME Importance (Aggregated)")
plt.tight_layout()
plt.show()
```
