```python
def get_sample_ids(df, exclude_cols):
    return [col for col in df.columns if col not in exclude_cols]
```

```python
luad_cna_samples = get_sample_ids(luad_cna, [ 'Entrez_Gene_Id'])
luad_meth_samples = get_sample_ids(luad_dna, ['ENTITY_STABLE_ID', 'DESCRIPTION', 'TRANSCRIPT_ID'])
```

```python
lusc_cna_samples = get_sample_ids(lusc_cna, [ 'Entrez_Gene_Id'])
lusc_meth_samples = get_sample_ids(lusc_dna, ['ENTITY_STABLE_ID',  'DESCRIPTION', 'TRANSCRIPT_ID'])
```

```python
luad_matched = sorted(set(luad_cna_samples).intersection(luad_meth_samples))
lusc_matched = sorted(set(lusc_cna_samples).intersection(lusc_meth_samples))
```

```python
print(f"Number of matched LUAD samples: {len(luad_matched)}")
print(f"Number of matched LUSC samples: {len(lusc_matched)}")
```

```python
luad_df = pd.DataFrame({
    'SampleID': luad_matched,
    'Class': ['LUAD'] * len(luad_matched)
})
lusc_df = pd.DataFrame({
    'SampleID': lusc_matched,
    'Class': ['LUSC'] * len(lusc_matched)
})

samples_df = pd.concat([luad_df, lusc_df])
```

```python
samples_df.head()
```

```python
samples_df.shape
```

```python
lusc_cna_features = lusc_cna[['Hugo_Symbol'] + lusc_matched]
luad_cna_features = luad_cna[['Hugo_Symbol'] + luad_matched]
lusc_dna_features = lusc_dna[['NAME'] + lusc_matched]
luad_dna_features = luad_dna[['NAME'] + luad_matched]
```

```python
cna_features=pd.concat([lusc_cna_features,luad_cna_features],axis=1)
dna_features=pd.concat([lusc_dna_features,luad_dna_features],axis=1)
```

```python
cna_features.head(10)
```

```python
dna_features.head(10)
```

```python
cna_features=cna_features.drop(columns='Hugo_Symbol')
cna_features.shape
```

```python
dna_features=dna_features.drop(columns='NAME')
dna_features.shape
```

```python
#Median Imputation 
from sklearn.impute import SimpleImputer

median_imp=SimpleImputer(strategy='median')
cna_imp=pd.DataFrame(
    median_imp.fit_transform(cna_features),
    index=cna_features.index,
    columns=cna_features.columns
)
cna_imp.isna().sum()
```

```python
dna_imp=pd.DataFrame(
    median_imp.fit_transform(dna_features),
    index=dna_features.index,
    columns=dna_features.columns
)
dna_imp.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
cna_z=pd.DataFrame(
    scaler.fit_transform(cna_imp),
    index=cna_imp.index,
    columns=cna_imp.columns
)
```

```python
beta = dna_imp.clip(1e-6, 1 - 1e-6)
m_values = np.log2(beta / (1 - beta))
dna_z = pd.DataFrame(
    scaler.fit_transform(m_values),
    index=m_values.index,
    columns=m_values.columns
)
```

```python
print(cna_z.shape)
print(dna_z.shape)
```

```python
features=pd.concat([cna_z, dna_z],axis=0)
```

```python
#Check for outliers
t_sample = features.median(axis=1)
q75 = features.quantile(0.75, axis=1)
q25 = features.quantile(0.25, axis=1)
iqr_sample = q75 - q25
```

```python
# Define global thresholds based on sample medians distribution
global_median = t_sample.median()
global_iqr = t_sample.quantile(0.75) - t_sample.quantile(0.25)
lower_bound = global_median - 3 * global_iqr
upper_bound = global_median + 3 * global_iqr
```

```python
outlier_samples = t_sample[(t_sample < lower_bound) | (t_sample > upper_bound)].index.tolist()
print("No. of outliers:", len(outlier_samples))
```

```python
labels=samples_df
labels.head(10)
```

```python
combined_features_t=combined_features.T
combined_features_t.head(10)
```

```python
X=combined_features_t.copy()
y=labels
```

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0.01)
X_var = pd.DataFrame(
    vt.fit_transform(X),
    index=X.index,
    columns=X.columns[vt.get_support()]
)
print(f"After variance filter: {X_var.shape[1]} features")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_var)
print(f"PCA output shape: {X_pca.shape}")
```

```python
print(X_pca)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
```

```python
if not hasattr(X_pca, 'index'):
    X = pd.DataFrame(X_pca, index=labels.index, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
else:
    X = X_pca.copy()
```

```python
y_enc = y['Class'].map({'LUAD':0, 'LUSC':1})
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y_enc, test_size=0.3, random_state=42, stratify=y_enc
)
```

```python
print(f"Variance explained by components: {np.sum(pca.explained_variance_ratio_)}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
logreg=LogisticRegression(max_iter=1000, solver='lbfgs',random_state=42)
```

```python
scores = cross_val_score(logreg, X_train, y_train, cv=5, scoring='accuracy')

# Output results
print("âœ… Logistic Regression CV scores:", scores)
print(f"âœ… Mean Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred = logreg.predict(X_test)
y_proba =logreg.predict_proba(X_test)[:, 1]
```

```python
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Test ROC AUC:", roc_auc_score(y_test, y_proba))
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=['LUAD','LUSC']))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from lime.lime_tabular import LimeTabularExplainer

explainer_lime = LimeTabularExplainer(
    training_data   = X_train,
    feature_names   = genes_names,
    class_names     = ['LUSC', 'LUAD'],
    mode            = 'classification'
)
```

```python
i = 0  # index of the test sample you want to inspect
exp = explainer_lime.explain_instance(
    data_row    = X_test[i],
    predict_fn  = logreg.predict_proba,     # your Keras model returns probability vector
    num_features= 10                 # top 10 genes for this sample
)
```

```python
feat, weight = zip(*exp.as_list(label=1))
plt.figure(figsize=(6,4))
plt.barh(feat[::-1], weight[::-1], color='coral')
plt.title(f"LIME Local Explanation (sample #{i})")
plt.xlabel("LIME weight")
plt.tight_layout()
plt.show()
```

```python
n_agg = 50  # number of random test samples to aggregate
agg_weights = np.zeros(len(genes_names))

# For each random sample, get LIME weights for class 1
idxs = np.random.choice(range(X_test.shape[0]), n_agg, replace=False)
for j in idxs:
    exp_j = explainer_lime.explain_instance(
        X_test[j],
        logreg.predict_proba,
        num_features=len(genes_names)   # get all gene weights
    )
    # as_list returns only the top num_features; instead grab .as_map()
    # .as_map()[class_id] is list of (feature_index, weight)
    local_map = dict(exp_j.as_map()[1])
    for feat_idx, w in local_map.items():
        agg_weights[feat_idx] += abs(w)

# average
agg_weights /= n_agg

# pick top 10 genes
top_idx   = np.argsort(agg_weights)[-10:][::-1]
top_genes = [str(genes_names[i]) for i in top_idx]
top_vals  = agg_weights[top_idx]

# â”€â”€ E) Plot global topâ€‘10 bar chart â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
plt.figure(figsize=(8,5))
plt.barh(top_genes[::-1], top_vals[::-1], color='teal')
plt.xlabel("Mean |LIME weight| over 50 samples")
plt.title("TopÂ 10 Genes by LIME Importance (Aggregated)")
plt.tight_layout()
plt.show()
```
