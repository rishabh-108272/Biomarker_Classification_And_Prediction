```python
from lifelines import CoxPHFitter
from lifelines import KaplanMeierFitter
import pandas as pd
```

```python
diff_ppi = pd.read_csv("./differential_ppi.csv")
```

```python
# Select top 20 PPIs
top_ppi = diff_ppi.sort_values("Corr_Diff", ascending=False).head(20)
print("Selected PPIs:")
print(top_ppi[['Protein1', 'Protein2', 'Corr_Diff']])
```

```python
luad_clinical_data=pd.read_csv('./data_clinical_patient_luad.txt', sep="\t", comment="#", skiprows=3)
lusc_clinical_data=pd.read_csv('./data_clinical_patient_lusc.txt', sep="\t", comment="#", skiprows=3)
```

```python
luad_survival = luad_clinical_data[['PATIENT_ID', 'OS_MONTHS', 'OS_STATUS']].copy()
luad_survival['subtype'] = 0  # LUAD
lusc_survival = lusc_clinical_data[['PATIENT_ID', 'OS_MONTHS', 'OS_STATUS']].copy()
lusc_survival['subtype'] = 1  # LUSC
```

```python
luad_survival['event'] = luad_survival['OS_STATUS'].map({'0:LIVING': 0, '1:DECEASED': 1})
lusc_survival['event'] = lusc_survival['OS_STATUS'].map({'0:LIVING': 0, '1:DECEASED': 1})
```

```python
survival_data = pd.concat([luad_survival, lusc_survival], axis=0)
survival_data = survival_data.rename(columns={'OS_MONTHS': 'time', 'PATIENT_ID': 'sample_id'})
survival_data = survival_data[['sample_id', 'time', 'event', 'subtype']].dropna()
```

```python
luad_rppa_t=pd.read_csv('./luad_rppa_t.csv')
lusc_rppa_t=pd.read_csv('./lusc_rppa_t.csv')

luad_rppa=luad_rppa_t
lusc_rppa=lusc_rppa_t
```

```python
luad_rppa.head()
```

```python
lusc_rppa.head()
```

```python
luad_rppa = luad_rppa.rename(columns={'Unnamed: 0': 'Composite.Element.REF'})
```

```python
lusc_rppa = lusc_rppa.rename(columns={'Unnamed: 0': 'Composite.Element.REF'})
```

```python
luad_rppa.index = luad_rppa['Composite.Element.REF'].str.slice(0, 12)  # e.g., TCGA-05-4244
lusc_rppa.index = lusc_rppa['Composite.Element.REF'].str.slice(0, 12)
rppa_data = pd.concat([luad_rppa, lusc_rppa], axis=0)
rppa_data=rppa_data.drop(columns=['Composite.Element.REF'])
```

```python
rppa_data.head()
```

```python
# Create PPI features (expression product)
features = pd.DataFrame(index=rppa_data.index)
for _, row in top_ppi.iterrows():
    p1, p2 = row['Protein1'], row['Protein2']
    if p1 in rppa_data.columns and p2 in rppa_data.columns:
        feature_name = f"{p1}_{p2}"
        features[feature_name] = rppa_data[p1] * rppa_data[p2]
    else:
        print(f"Warning: {p1} or {p2} not in RPPA data.")
```

```python
# Merge with survival data
data = survival_data.set_index('sample_id').join(features, how='inner')
data = data.dropna()
```

```python
print("\nMerged Data Shape:", data.shape)
print("Merged Data Sample:")
data.head()
```

```python
cox_data = data[['time', 'event', 'subtype'] + features.columns.tolist()]
cph = CoxPHFitter(penalizer=0.1)
try:
    cph.fit(cox_data, duration_col='time', event_col='event', strata="subtype")
except Exception as e:
    print(f"Error fitting Cox model: {e}")
    exit()
```

```python
print("\nCox Regression Summary:")
cph.print_summary(decimals=3)

# Save coefficients
results = cph.summary[['coef', 'exp(coef)', 'p']]
results.to_csv("cox_ppi_results.csv")
print("\nSaved coefficients to cox_ppi_results.csv")
```

```python
from matplotlib import pyplot as plt 
if features.columns.size > 0:
    top_feature = features.columns[0]
    km_data = data[[top_feature, 'time', 'event']].copy()
    km_data['group'] = km_data[top_feature] > km_data[top_feature].median()

    kmf = KaplanMeierFitter()
    plt.figure(figsize=(8, 6))
    for group in [True, False]:
        mask = km_data['group'] == group
        kmf.fit(km_data[mask]['time'], km_data[mask]['event'], 
                label=f"{top_feature} {'High' if group else 'Low'}")
        kmf.plot_survival_function()

    plt.title(f"Kaplan-Meier Survival by {top_feature}")
    plt.savefig("kaplan_meier_ppi.png")
    plt.close()
else:
    print("No valid PPI features for Kaplan-Meier plot.")
```

```python
os_survival = pd.concat([luad_clinical_data[['PATIENT_ID', 'OS_MONTHS', 'OS_STATUS', 'subtype']], 
                         lusc_clinical_data[['PATIENT_ID', 'OS_MONTHS', 'OS_STATUS', 'subtype']]], axis=0)
os_survival = os_survival.rename(columns={'OS_MONTHS': 'time', 'PATIENT_ID': 'sample_id', 'OS_STATUS':'event'})
os_survival = os_survival.dropna()

dfs_survival = pd.concat([luad_clinical_data[['PATIENT_ID', 'DFS_MONTHS', 'DFS_STATUS', 'subtype']], 
                          lusc_clinical_data[['PATIENT_ID', 'DFS_MONTHS', 'DFS_STATUS', 'subtype']]], axis=0)
dfs_survival = dfs_survival.rename(columns={'DFS_MONTHS': 'time',  'PATIENT_ID': 'sample_id', 'DFS_STATUS':'event'})
dfs_survival = dfs_survival.dropna()

print("OS Survival Data Shape:", os_survival.shape)
print("DFS Survival Data Shape:", dfs_survival.shape)
```

```python
def cox_predict(data, features, endpoint):
    if data.shape[0] < 20:
        print(f"Error: {endpoint} data too small ({data.shape[0]} samples). Skipping.")
        return None
    
    cox_data = data[['time', 'event', 'subtype'] + features.columns.tolist()]
    cph = CoxPHFitter()
    try:
        cph.fit(cox_data, duration_col='time', event_col='event')
    except Exception as e:
        print(f"Error fitting {endpoint} Cox model: {e}")
        return None
    
    # Predict survival probabilities at 12, 36, 60 months
    # Predict survival probabilities at 12, 36, 60 months
    time_points = [12, 36, 60]
    # Ensure correct columns are passed
    covariates = cox_data[features.columns.tolist() + ['subtype']]
    survival_prob = cph.predict_survival_function(covariates, times=time_points)
    # survival_prob has shape (len(time_points), n_samples); transpose to (n_samples, len(time_points))
    survival_prob = survival_prob.T
    survival_prob.index = cox_data.index
    survival_prob.index.name = 'sample_id'
    survival_prob.columns = [f"{endpoint}_prob_{t}m" for t in time_points]
    
    # Predict risk scores
    risk_scores = cph.predict_log_partial_hazard(cox_data[features.columns.tolist() + ['subtype']])
    risk_scores = pd.Series(risk_scores, index=cox_data.index, name=f"{endpoint}_risk_score")

    # Save results
    results = cph.summary[['coef', 'exp(coef)', 'p']]
    results.to_csv(f"cox_{endpoint}_results.csv")
    
    # Kaplan-Meier for top feature
    if features.columns.size > 0:
        top_feature = features.columns[0]
        km_data = data[[top_feature, 'time', 'event']].copy()
        km_data['group'] = km_data[top_feature] > km_data[top_feature].median()
        
        kmf = KaplanMeierFitter()
        plt.figure(figsize=(8, 6))
        for group in [True, False]:
            mask = km_data['group'] == group
            if mask.sum() > 0:
                kmf.fit(km_data[mask]['time'], km_data[mask]['event'], 
                        label=f"{top_feature} {'High' if group else 'Low'}")
                kmf.plot_survival_function()
        
        plt.title(f"Kaplan-Meier {endpoint} by {top_feature}")
        plt.savefig(f"kaplan_meier_{endpoint}.png")
        plt.close()
    
    # Save predictions
    predictions = pd.concat([survival_prob, risk_scores], axis=1)
    predictions.to_csv(f"{endpoint}_predictions.csv")
    predictions = pd.concat([survival_prob, risk_scores], axis=1)
    predictions.to_csv(f"{endpoint}_predictions.csv")
    
    # Print summary
    print(f"\n{endpoint} Cox Regression Summary:")
    cph.print_summary(decimals=3)
    print(f"{endpoint} Concordance:", cph.concordance_index_)
    
    return predictions
```

```python
# Predict OS
os_predictions = cox_predict(os_data, features, "OS")
if os_predictions is not None:
    print("\nOS Predictions Sample:")
    print(os_predictions.head())
```

```python
# Function to validate Cox model
def cox_validate(data, features, endpoint):
    if data.shape[0] < 20:
        print(f"Error: {endpoint} data too small ({data.shape[0]} samples). Skipping.")
        return None, None
    
    # Train/test split
    train, test = train_test_split(data, test_size=0.3, random_state=42)
    print(f"{endpoint} Train Shape:", train.shape)
    print(f"{endpoint} Test Shape:", test.shape)
    
    # Define covariates
    covariates = ['subtype'] + features.columns.tolist()
    cox_columns = ['time', 'event'] + covariates
    
    # Fit model on training data
    cph = CoxPHFitter()
    try:
        cph.fit(train[cox_columns], duration_col='time', event_col='event')
    except Exception as e:
        print(f"Error fitting {endpoint} Cox model: {e}")
        return None, None
        # Save model
    model_file = f"cox_{endpoint}_train_model.joblib"
    joblib.dump(cph, model_file)
    print(f"Saved {endpoint} model to {model_file}")
    
    # Save feature names
    pd.Series(covariates).to_csv(f"cox_{endpoint}_train_features.csv", index=False)
    
    # Predict on test set
    time_points = [12, 36, 60]
    test_prob = cph.predict_survival_function(test[covariates], times=time_points).T
    test_prob.index = test.index
    test_prob.index.name = 'sample_id'
    test_prob.columns = [f"{endpoint}_prob_{t}m" for t in time_points]
    
    # Predict risk scores
    test_risk = cph.predict_log_partial_hazard(test[covariates])
    test_risk = pd.Series(test_risk, index=test.index, name=f"{endpoint}_risk_score")
    
    # Save test predictions
    test_predictions = pd.concat([test_prob, test_risk], axis=1)
    test_predictions.to_csv(f"test_{endpoint}_predictions.csv")
    
    # Print summary
    print(f"\n{endpoint} Cox Regression Summary:")
    cph.print_summary(decimals=3)
    print(f"{endpoint} Concordance:", cph.concordance_index_)
    print(f"\n{endpoint} Test Predictions Sample:")
    print(test_predictions.head())
    
    return test_predictions, cph
```

```python
import joblib 
from sklearn.model_selection import train_test_split
# Validate OS
os_test_predictions, os_cph = cox_validate(os_data, features, "OS")
if os_test_predictions is not None:
    print("\nOS Test Predictions Saved to test_OS_predictions.csv")

# Validate DFS
dfs_test_predictions, dfs_cph = cox_validate(dfs_data, features, "DFS")
if dfs_test_predictions is not None:
    print("\nDFS Test Predictions Saved to test_DFS_predictions.csv")
```
