<h1>Correlation Analysis for PPI inference</h1>

```python
from scipy.stats import pearsonr
def compute_correlation_matrix(data):
    proteins = data.columns
    n_proteins = len(proteins)
    corr_matrix = np.zeros((n_proteins, n_proteins))
    pval_matrix = np.zeros((n_proteins, n_proteins))
    
    for i in range(n_proteins):
        for j in range(i + 1, n_proteins):
            corr, pval = pearsonr(data[proteins[i]], data[proteins[j]])
            corr_matrix[i, j] = corr
            corr_matrix[j, i] = corr
            pval_matrix[i, j] = pval
            pval_matrix[j, i] = pval
    
    return pd.DataFrame(corr_matrix, index=proteins, columns=proteins), pd.DataFrame(pval_matrix, index=proteins, columns=proteins)
```

```python
luad_corr, luad_pval = compute_correlation_matrix(luad_rppa_t)
lusc_corr, lusc_pval = compute_correlation_matrix(lusc_rppa_t)
```

```python
print(luad_rppa_t.shape)
print(lusc_rppa_t.shape)
```

```python
from statsmodels.stats.multitest import multipletests
# Apply FDR correction
luad_pval_flat = luad_pval.values[np.triu_indices_from(luad_pval, k=1)]
lusc_pval_flat = lusc_pval.values[np.triu_indices_from(lusc_pval, k=1)]
_, luad_adj_pval_flat, _, _ = multipletests(luad_pval_flat, method="fdr_bh")
_, lusc_adj_pval_flat, _, _ = multipletests(lusc_pval_flat, method="fdr_bh")
```

```python
# Reconstruct adjusted p-value matrices
luad_adj_pval = pd.DataFrame(np.zeros_like(luad_pval), index=luad_pval.index, columns=luad_pval.columns)
lusc_adj_pval = pd.DataFrame(np.zeros_like(lusc_pval), index=lusc_pval.index, columns=lusc_pval.columns)
triu_indices = np.triu_indices_from(luad_pval, k=1)
luad_adj_pval.values[triu_indices] = luad_adj_pval_flat
luad_adj_pval.values.T[triu_indices] = luad_adj_pval_flat
lusc_adj_pval.values[triu_indices] = lusc_adj_pval_flat
lusc_adj_pval.values.T[triu_indices] = lusc_adj_pval_flat
```

```python
# Select significant correlations (|corr| > 0.5, adj_pval < 0.05)
luad_edges = []
lusc_edges = []
for i, prot1 in enumerate(luad_corr.index):
    for j, prot2 in enumerate(luad_corr.columns[i+1:]):
        j = j + i + 1
        if abs(luad_corr.iloc[i, j]) > 0.5 and luad_adj_pval.iloc[i, j] < 0.05:
            luad_edges.append((prot1, prot2, luad_corr.iloc[i, j]))
        if abs(lusc_corr.iloc[i, j]) > 0.5 and lusc_adj_pval.iloc[i, j] < 0.05:
            lusc_edges.append((prot1, prot2, lusc_corr.iloc[i, j]))
```

```python
# Save correlation results
luad_corr.to_csv("luad_ppi_correlations.csv")
lusc_corr.to_csv("lusc_ppi_correlations.csv")
luad_adj_pval.to_csv("luad_ppi_pvalues.csv")
lusc_adj_pval.to_csv("lusc_ppi_pvalues.csv")
```

```python
# Step 3: Map Correlations to Known PPI Networks
# Use stringdb package
import stringdb
proteins = list(set(luad_rppa.index))  # 198 unique proteins
try:
    # Map RPPA protein names to STRING IDs
    string_ids = stringdb.get_string_ids(proteins, species=9606)
    print("Mapped proteins to STRING IDs:", string_ids.shape)
    
    # Get PPI network from STRING
    network_df = stringdb.get_network(string_ids['queryItem'], required_score=700)
    print("STRING network edges:", network_df.shape)
    
    # Convert STRING network to edge list (use preferredName for mapping back to RPPA names)
    string_edges = [
        (row['preferredName_A'], row['preferredName_B'], row['score'])
        for _, row in network_df.iterrows()
    ]
except Exception as e:
    print(f"Error querying STRING API: {e}")
    print("Falling back to mock STRING edges")
```

```python
# Intersect with correlation-based edges
luad_ppi = [(p1, p2, corr) for p1, p2, corr in luad_edges 
            if (p1, p2) in [(e[0], e[1]) for e in string_edges] or (p2, p1) in [(e[0], e[1]) for e in string_edges]]
lusc_ppi = [(p1, p2, corr) for p1, p2, corr in lusc_edges 
            if (p1, p2) in [(e[0], e[1]) for e in string_edges] or (p2, p1) in [(e[0], e[1]) for e in string_edges]]
```

```python
# Save PPI lists
pd.DataFrame(luad_ppi, columns=["Protein1", "Protein2", "Correlation"]).to_csv("luad_ppi_edges.csv", index=False)
pd.DataFrame(lusc_ppi, columns=["Protein1", "Protein2", "Correlation"]).to_csv("lusc_ppi_edges.csv", index=False)
```

```python
# Step 4: Differential PPI Network Analysis
diff_ppi = []
for p1, p2, luad_corr in luad_ppi:
    lusc_match = [(p1_, p2_, corr) for p1_, p2_, corr in lusc_ppi if (p1_, p2_) == (p1, p2) or (p2_, p1_) == (p1, p2)]
    lusc_corr = lusc_match[0][2] if lusc_match else 0
    corr_diff = abs(luad_corr - lusc_corr)
    if corr_diff > 0.3:
        diff_ppi.append((p1, p2, luad_corr, lusc_corr, corr_diff))

for p1, p2, lusc_corr in lusc_ppi:
    if not any((p1_, p2_) == (p1, p2) or (p2_, p1_) == (p1, p2) for p1_, p2_, _, _, _ in diff_ppi):
        luad_match = [(p1_, p2_, corr) for p1_, p2_, corr in luad_ppi if (p1_, p2_) == (p1, p2) or (p2_, p1_) == (p1, p2)]
        luad_corr = luad_match[0][2] if luad_match else 0
        corr_diff = abs(lusc_corr - luad_corr)
        if corr_diff > 0.3:
            diff_ppi.append((p1, p2, luad_corr, lusc_corr, corr_diff))
```

```python
# Save differential PPI
diff_ppi_df = pd.DataFrame(diff_ppi, columns=["Protein1", "Protein2", "LUAD_Corr", "LUSC_Corr", "Corr_Diff"])
diff_ppi_df.to_csv("differential_ppi.csv", index=False)
```

```python
# Step 7: Validation and Integration
# Split samples for correlation stability
luad_train = luad_rppa_t.sample(frac=0.7, random_state=42)
luad_test = luad_rppa_t.drop(luad_train.index)
lusc_train = lusc_rppa_t.sample(frac=0.7, random_state=42)
lusc_test = lusc_rppa_t.drop(lusc_train.index)

luad_corr_train, _ = compute_correlation_matrix(luad_train)
lusc_corr_train, _ = compute_correlation_matrix(lusc_train)
```

```python

# Step 1: Define correlation computation function
def compute_correlation_matrix(data):
    proteins = data.columns
    n_proteins = len(proteins)
    corr_matrix = np.zeros((n_proteins, n_proteins))
    pval_matrix = np.zeros((n_proteins, n_proteins))
    
    for i in range(n_proteins):
        for j in range(i + 1, n_proteins):
            corr, pval = pearsonr(data[proteins[i]], data[proteins[j]])
            corr_matrix[i, j] = corr
            corr_matrix[j, i] = corr
            pval_matrix[i, j] = pval
            pval_matrix[j, i] = pval
    
    return pd.DataFrame(corr_matrix, index=proteins, columns=proteins), pd.DataFrame(pval_matrix, index=proteins, columns=proteins)

# Step 2: Load precomputed correlation matrices from CSV
luad_corr = pd.read_csv('./luad_ppi_correlations.csv', index_col=0)
lusc_corr = pd.read_csv('./lusc_ppi_correlations.csv', index_col=0)

# Step 3: Drop any extra columns not part of the index
luad_corr = luad_corr.loc[:, luad_corr.columns.intersection(luad_corr.index)]
luad_corr = luad_corr.loc[luad_corr.index.intersection(luad_corr.columns)]
lusc_corr = lusc_corr.loc[:, lusc_corr.columns.intersection(lusc_corr.index)]
lusc_corr = lusc_corr.loc[lusc_corr.index.intersection(lusc_corr.columns)]

# Step 4: Split RPPA samples
luad_train = luad_rppa_t.sample(frac=0.7, random_state=42)
luad_test = luad_rppa_t.drop(luad_train.index)
lusc_train = lusc_rppa_t.sample(frac=0.7, random_state=42)
lusc_test = lusc_rppa_t.drop(lusc_train.index)

# Step 5: Compute correlation matrices on train data
luad_corr_train, _ = compute_correlation_matrix(luad_train)
lusc_corr_train, _ = compute_correlation_matrix(lusc_train)

# Step 6: Align columns and index between original and train correlation matrices
luad_corr_train = luad_corr_train.loc[luad_corr.index, luad_corr.columns]
lusc_corr_train = lusc_corr_train.loc[lusc_corr.index, lusc_corr.columns]

# Step 7: Compute upper triangle and correlation stability
luad_upper = luad_corr.values[np.triu_indices_from(luad_corr, k=1)]
luad_train_upper = luad_corr_train.values[np.triu_indices_from(luad_corr_train, k=1)]
luad_corr_stability = np.corrcoef(luad_upper, luad_train_upper)[0, 1]

lusc_upper = lusc_corr.values[np.triu_indices_from(lusc_corr, k=1)]
lusc_train_upper = lusc_corr_train.values[np.triu_indices_from(lusc_corr_train, k=1)]
lusc_corr_stability = np.corrcoef(lusc_upper, lusc_train_upper)[0, 1]

# Step 8: Display results
print(f"LUAD Correlation Stability: {luad_corr_stability:.3f}")
print(f"LUSC Correlation Stability: {lusc_corr_stability:.3f}")
```
